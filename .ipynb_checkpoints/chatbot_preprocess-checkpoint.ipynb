```python
import nltk
import random
import numpy as np
import json
import pickle
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()

with open("intents.json") as f:
    data = json.load(f)
```

```python
words = []
labels = []
docs = []
```

```python
for intent in data["intents"]:
    
    for pattern in intent["patterns"]:

        token_words = nltk.word_tokenize(pattern)       # Tokenizing each word

        # docs contains pair of tokenized words and the corresponding tags
        # Eg : ["hi"], 'greeting'

        docs.append([token_words,intent["tag"]])
        
        words.extend(token_words)       # we do not need linked lists as tokenize creates list by default

    if intent["tag"] not in labels:
        labels.append(intent["tag"])        # Adding labels
```

```python
words = [stemmer.stem(word.lower()) for word in words if word != ["?",","]]
words = sorted(list(set(words)))        # using set to remove the duplicates

labels = sorted(list(set(labels)))

# 28 docs
# 4 labels
# 26 uniques tokenized words
```

```python
training = []
result_null = [0] * len(labels)
for doc in docs:
    # initializing bag of words
    bow = []
    # list of tokenized words for the pattern
    pattern_words = doc[0]
    # stemming each word
    pattern_words = [stemmer.stem(word.lower()) for word in pattern_words]

    # create our BOW array with 1, if word match found in current pattern
    for w in words:
        if w in pattern_words:
            bow.append(1)
        else:
            bow.append(0)

    # output is a '0' for each tag and '1' for current tag (for each pattern)
    result_row = list(result_null)         # creating copy 
    result_row[labels.index(doc[1])] = 1
    print(result_row)
    training.append([bow, result_row])
```

```python
# shuffle our features and turn into np.array
random.shuffle(training)
training = np.array(training)
# print(training)
# X -> patterns, y -> intents
train_x = list(training[:,0])
train_y = list(training[:,1])
print("Training data created")
print(labels)
```

```python
model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))
# Compile model. Stochastic gradient descent with Nesterov accelerated gradient gives good results for this model
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
#fitting and saving the model 
hist = model.fit(np.array(train_x), np.array(train_y), epochs=200, batch_size=5, verbose=1)
model.save('chatbot_model.h5', hist)
print("model created")
```

```python
pickle.dump(words,open('words.pkl','wb'))
pickle.dump(labels,open('labels.pkl','wb'))
```
