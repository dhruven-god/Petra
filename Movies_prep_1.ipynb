```python
import pandas as pd
import numpy as np
%matplotlib inline

import warnings; warnings.simplefilter('ignore')
df1 = pd.read_csv('movies_metadata.csv')

df1.head()
```

```python
df2 = pd.read_csv("credits.csv")

df2.head()
```

```python
df1["cast"] = df2["cast"]
df1["crew"] = df2["crew"]

df = df1
```

```python
df.head()
```

```python
df['original_language'].unique()
```

```python
df.dropna(inplace=True)     # dropping null values
```

```python
df.isnull().sum()
```

```python
from ast import literal_eval
def clean(genre):
    
    lst = []
    for i in literal_eval(genre):         # Converting string object into list so that it is iteratable
        lst.append(i['name'])
        
    return lst
        
```

```python
df ['genres'] = df['genres'].apply(clean)
df['cast'] = df['cast'].apply(clean)

df['title'] = df['title'].str.lower()      # lowering all title names
df.head()
```

```python
df = df[["title","genres","overview","cast","crew"]]   # picking up only necessary columns
df.head()
```

```python
def director(text):
    L = []
    for i in literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
df['crew'] = df['crew'].apply(director)   # getting name of director to get similar movies by him
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])   # Removing spacing between single word
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])   # Removing spacing between single word# Removing spacing between single word
```

```python
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
df
```

```python
df['summary'] = df['overview'] + df['genres'] + df['cast'] + df['crew']
df['summary'] = df['summary'].apply(lambda x: " ".join(x))
```
